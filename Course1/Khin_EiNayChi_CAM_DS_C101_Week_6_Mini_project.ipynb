Frequency based on purchases guides a business in the effectiveness of target marketing campaigns and how to target less active customers.

- The **average unit cost** indicates if the customer prefers low cost or high cost items. This is related to the profitability of purchases. Customers buying products with a higher average unit cost price should be targeted differently. Customer segmentation assists in identifying these customers.

```python
df_cus[df_cus.isnull().any(axis=1)]
```

```python
#Dealing with Nulls
geo_cols = ['City', 'State_Province', 'Postal_Code']
for col in geo_cols:
    df_cus[col] = df_cus[col].fillna('Unknown')
```

```python
#DateTime conversion for consistency and further calculation
df_cus['Order_Date'] = pd.to_datetime(df_cus['Order_Date'])
df_cus['Customer_BirthDate'] = pd.to_datetime(df_cus['Customer_BirthDate'])
df_cus['Delivery_Date'] = pd.to_datetime(df_cus['Delivery_Date'])

df_cus[['Order_Date', 'Customer_BirthDate', 'Delivery_Date']].dtypes
```

```python
df_cus[geo_cols].isnull().sum()
```

```python
#Formatting Currency Columns
currency_cols = ['Total Revenue', 'Unit Cost', 'Profit']

for col in currency_cols:
    df_cus[col] = (df_cus[col].astype(str) #change into string first
    .str.replace('(', '-', regex= False) #negative values
    .str.replace(r'[$,)]', '', regex=True)
    )
    df_cus[col] = pd.to_numeric(df_cus[col], errors='coerce') #numeric conversion
print(df_cus[currency_cols].dtypes)
```

```python
df_cus.head()
```

```python
df_cus[currency_cols].isnull().sum()
```

```python
#confirmation of all features are not null.
df_cus.isnull().sum()
```

```python
#Prepartion for guest and returned customer
df_cus['unified_id'] = np.where(df_cus['Loyalty Num'] == 0,
                                df_cus['Customer ID'],
                                df_cus['Loyalty Num'])

df_cus[['unified_id', 'Customer ID', 'Loyalty Num']]
```

```python
#Creating five new columns for each data entry
today = pd.Timestamp.now().normalize()

customer_df = df_cus.groupby('unified_id').agg(
    #Recency: days since the last purchase
    Recency = ('Order_Date', lambda x: (today - x.max()).days),
    #Frequency
    Frequency=('Order ID', 'nunique'),
    # CLV: total sum of revenue
    CLV = ('Total Revenue', 'sum'),
    # Avg unit cost
    Avg_Unit_Cost = ('Unit Cost', 'mean'),
    # Age
    Age = ('Customer_BirthDate', lambda x: int((today - x.min()). days / 365))

).reset_index()
```

```python
#Treating outliers on the logged data
cols_to_filter=['CLV', 'Frequency', 'Recency', 'Avg_Unit_Cost']
initial_rows = len(customer_df)

for col in cols_to_filter:
  Q1 = customer_df[col].quantile(0.25)
  Q3 = customer_df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  #filtering outliers/ remove
  customer_df = customer_df[(customer_df[col] >= lower_bound) & (customer_df[col] <= upper_bound)]
```

```python
rows_removed = initial_rows - len(customer_df)
rows_removed
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
customer_df_scaled = pd.DataFrame(scaler.fit_transform(customer_df[cols]), columns=cols, index=customer_df.index)
customer_df_scaled.max()
```

```python
#Dimensional Reduction
#PCA
#Analysis Code
from sklearn.decomposition import PCA
pca = PCA()
pca_results = pca.fit_transform(customer_df_scaled)

variance = np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10, 5))
plt.step(range(1, len(variance) + 1), variance, where='mid')
plt.ylabel('Cumulative Explained Variance')
plt.xlabel('No. of Principal Components')
plt.title('Explained Variance by Principal Components')
plt.axhline(y=0.9, color='r', linestyle='--') #threshold line
plt.show()
```

```python
#PCA reduction
pca_final = PCA(n_components=3)
pca_results_final = pca_final.fit_transform(customer_df_scaled)
```

```python
pca_results_final.shape
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch
```

```python
#num of clusters
wcss = []
silhouette_avgs = []
k_range = range(2, 11)

for k in k_range:
  kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
  labels = kmeans.fit_predict(pca_results_final)
  wcss.append(kmeans.inertia_)
  silhouette_avgs.append(silhouette_score(pca_results_final, labels))

#Elbow
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, 'bo-')
plt.title('Elbow Method')
plt.show()
plt.close()

#Silhouette
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_avgs, 'ro-')
plt.title('Silhouette Scores')
plt.show()
plt.close()
```

```python
final_kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
cluster_labels = final_kmeans.fit_predict(pca_results_final)

customer_df['Cluster'] = cluster_labels
segment_profile = customer_df.drop(columns=['unified_id']).groupby('Cluster').mean()
segment_profile
```
